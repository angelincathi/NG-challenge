#### 2. Diberikan 3 buah vector 3 dimensi:


```python
#membuat vektor A, B, C
A = np.array([17, 22, 19])
B = np.array([10, 20, 11])
C = np.array([5, 12, 9])
print(f'A = {A}\nB = {B}\nC = {C}')
```

```python
#Hitunglah A + B
AplusB = A + B

#Hitunglah B-C
BminusC = B-C

#A dot C
AdotC = A@C

#A x B
AcrossB = A*B

#norm A
Anorm = np.linalg.norm(A)

#Sudut antara vektor A dan B
cos_theta = (A@B)/(np.linalg.norm(A) * np.linalg.norm(B))
theta = np.arccos(cos_theta)

print(f'A + B = {AplusB}\nB - C = {BminusC}\nA dot C = {AdotC}\nA x B = {AcrossB}\nNorm A = {Anorm}\nSudut vektor A dan B = {theta}')
```

#### 3. Buatlah plot dari vektor berikut ini ke dalam bidang 2D : 𝒖=(2  5), 𝒗=(3  1)

```python
import matplotlib.pyplot as plt
u = np.array([2, 5])
v = np.array([3, 1])

def plot_vector2d(vector2d, origin=[0, 0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
              head_width=0.2, head_length=0.2, length_includes_head=True,
              **options)

plot_vector2d(u, color='r')
plot_vector2d(v, color='b')
plt.axis([-1, 6, -1, 6])
plt.title('Vektor 2 Dimensi')
plt.xlabel('Sumbu-X')
plt.ylabel('Sumbu-Y')

plt.grid()
plt.show()
```
